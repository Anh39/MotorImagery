```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# Example signal: a mix of low and high frequency components
fs = 500  # Sampling frequency in Hz
t = np.linspace(0, 1, fs, endpoint=False)  # Time vector
low_freq = 5  # Low frequency in Hz
high_freq = 50  # High frequency in Hz

# Mixed signal with low and high frequency components
x = np.sin(2 * np.pi * low_freq * t) + 0.5 * np.sin(2 * np.pi * high_freq * t)

# Design filters
low_cutoff = 10  # Low-pass cutoff frequency in Hz
high_cutoff = 20  # High-pass cutoff frequency in Hz
order = 4  # Filter order

# Low-pass filter
b_low, a_low = signal.butter(order, low_cutoff, btype='low', fs=fs)
low_passed = signal.filtfilt(b_low, a_low, x)

# High-pass filter
b_high, a_high = signal.butter(order, high_cutoff, btype='high', fs=fs)
high_passed = signal.filtfilt(b_high, a_high, x)

# Plot results
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, x, label='Original Signal')
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, low_passed, label='Low-Passed Signal', color='orange')
plt.title('Low-Passed Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, high_passed, label='High-Passed Signal', color='green')
plt.title('High-Passed Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```
